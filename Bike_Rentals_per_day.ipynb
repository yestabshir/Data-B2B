```python
# prompt: Using dataframe df2: matplotlib

import matplotlib.pyplot as plt

# Create a plot of the 'registered' column over time using the 'casual' column as the x-axis
plt.plot(df2['casual'], df2['registered'])

# Set the title and labels of the plot
plt.title('Bike Rentals Registered')
plt.xlabel('registered')
plt.ylabel('Number of Rentals')

# Rotate the x-axis labels for better readability (optional)
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
